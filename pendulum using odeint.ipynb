```python
from scipy.integrate import odeint
import numpy as np
```

```python
def pendulum(rhs,t,b,c):
    theta,omega=rhs
    drhsdt=[omega,-b*omega-c*np.sin(theta)]
    return drhsdt 
```

```python
b=0.25
c=5.0
```

```python
y0=[np.pi-0.1,0]
```

```python
t=np.linspace(0,10,101)
```

```python
sol = odeint(pendulum, y0, t, args=(b,c))
```

```python
import matplotlib.pyplot as plt
plt.plot(t,sol[:,0],'x',label='theta(t)')
plt.plot(t,sol[:,1],'y',label='omega(t)')
plt.legend(loc='best')
plt.xlabel('t')
plt.grid()
plt.show()
```
