```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
num_jumps=100
num_experiments=1000
x_array,y_array = np.zeros(num_jumps),np.zeros(num_jumps)
```

```python
def random_walk(x,y):
    random_num=np.random.random()
    if random_num <0.25:
        x,y=x-1,y #go left
    elif random_num <0.5:
        x,y=x+1,y #go_right
    elif random_num <0.75:
        x,y=x,y-1 #go_down
    else:
        x,y=x,y+1 #go_up
    return x,y
```

```python
data=np.zeros((num_experiments,num_jumps))
distance_sqr_sum=np.zeros(num_jumps)
jump_array=np.arange(0,num_jumps,1)

for i in range(num_experiments):
    x_array[0],y_array[0]=0,0
    for j in range(1,num_jumps):
        x_array[j],y_array[j]=random_walk(x_array[j-1],y_array[j-1])
        distance_sqr=x_array[j]**2+y_array[j]**2
        distance_sqr_sum[j] +=distance_sqr
        data[i,j]=np.sqrt(distance_sqr)
    plt.scatter(x_array,y_array,c=jump_array)
plt.colorbar()
plt.xlabel('x')
plt.ylabel('y')
```

```python
mean_sqr_dist=distance_sqr_sum/num_experiments #take horizontal line ,square x's and sum then divide with num_expts.

plt.plot(jump_array,mean_sqr_dist)
```
