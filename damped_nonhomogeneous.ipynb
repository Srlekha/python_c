```python
from scipy.integrate import odeint
import numpy as np
```

```python
def damped_nonhomogeneous(rhs,t,m,c,k,F_0,omega_0):
    y,v=rhs
    drhsdt=[v,-c/m*v-k/m*y-F_0*np.cos(omega_0*t)]
    return drhsdt 
```

```python
m=10
c=0
k=90
omega_0=2.8
F_0=5
```

```python
y0=[5,0]
```

```python
t=np.linspace(0,100,1000)
```

```python
sol = odeint(damped_nonhomogeneous, y0, t, args=(m,c,k,F_0,omega_0))
```

```python
import matplotlib.pyplot as plt
plt.plot(t,sol[:,0],label='y')
plt.plot(t,sol[:,1],label='v')
plt.legend(loc='best')
plt.xlabel('t')
plt.grid()
plt.show()
```

```python
def damped_nonhomogeneous2(rhs,t,m,c,k):
    y,v=rhs
    drhsdt=[v,-c/m*v-k/m*y-10*np.exp(-1.5*t)]
    return drhsdt 
```

```python
m=1
c=3
k=2.25
```

```python
sol1 = odeint(damped_nonhomogeneous2, y0, t, args=(m,c,k))
```

```python
import matplotlib.pyplot as plt
plt.plot(t,sol1[:,0],label='y')
plt.plot(t,sol1[:,1],label='v')
plt.legend(loc='best')
plt.xlabel('t')
plt.grid()
plt.show()
```

```python
c1=5
c2=7.5
```

```python
y_an_2=(c1+c2*t)*np.exp(-1.5*t)+(-5)*t**2 *(np.exp(-1.5*t))
```

```python
import matplotlib.pyplot as plt
plt.plot(t,sol1[:,0],label='y')
#plt.plot(t,sol1[:,1],label='v')
plt.plot(t,y_an_2,'--',label='analytical')
plt.legend(loc='best')
plt.xlabel('t')
plt.grid()
plt.show()
```
