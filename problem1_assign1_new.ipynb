Given,
Third condition: Assume the prey are infected by a disease and the infection spreads linearly killing the prey.
=> Hence, there should be a modification to prey equation ,an addition of (-theta*x) on RHS indicating 
the linear decrease of prey population

```python
from scipy.integrate import odeint
import numpy as np
```

```python
def abc(rhs,t,alpha,beta,gamma,delta,theta):
    x,y=rhs
    drhsdt=[alpha*x-beta*x*y-theta*x,delta*x*y-gamma*y]
    return drhsdt 
```

```python
alpha=4
beta=2
gamma=3
delta=4
theta=1.5
```

```python
y0=[5,5]
```

```python
t=np.linspace(0,100,10001)
```

```python
sol = odeint(abc, y0, t, args=(alpha,beta,gamma,delta,theta))
```

```python
import matplotlib.pyplot as plt
plt.plot(t,sol[:,0],'x',label='x(t)')
plt.plot(t,sol[:,1],'y',label='y(t)')
plt.legend(loc='best')
plt.xlabel('t')
plt.xlim([0,10])
plt.grid()
plt.show()
```

```python
plt.scatter(sol[:,1],sol[:,0],label='x(t) vs y(t) 5,5')
plt.legend(loc='best')
plt.show()
```
