```python
from scipy.integrate import odeint
import numpy as np
```

```python
def lodka_voltera(rhs,t,alpha,beta,gamma,delta):
    x,y=rhs
    drhsdt=[alpha*x-beta*x*y,delta*x*y-gamma*y]
    return drhsdt 
```

```python
alpha=4
beta=2
gamma=3
delta=4
```

```python
y0=[5,5]
```

```python
t=np.linspace(0,100,10001)
```

```python
sol = odeint(lodka_voltera, y0, t, args=(alpha,beta,gamma,delta))
```

```python
y1_0=[10,5]
```

```python
sol1 = odeint(lodka_voltera, y1_0, t, args=(alpha,beta,gamma,delta))
```

```python
y2_0=[5,10]
```

```python
sol2 = odeint(lodka_voltera, y2_0, t, args=(alpha,beta,gamma,delta))
```

```python
import matplotlib.pyplot as plt
plt.plot(t,sol[:,0],'x',label='x(t)')
plt.plot(t,sol[:,1],'y',label='y(t)')
plt.legend(loc='best')
plt.xlabel('t')
plt.xlim([0,10])
plt.grid()
plt.show()
```

```python
plt.scatter(sol[:,1],sol[:,0],label='x(t) vs y(t) 5,5')
plt.scatter(sol1[:,1],sol1[:,0],label='x(t) vs y(t) 10,5')
plt.scatter(sol2[:,1],sol2[:,0],label='x(t) vs y(t) 5,10')
plt.legend(loc='best')
plt.show()
```
