```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
def Jacobi_2D(sol,B):
    Niter,Nx,Ny=sol.shape
    for iter in range(1,Niter,1):
        for i in range(1,Nx-1,1):
            for j in range(1,Ny-1,1):
                sol[iter,i,j]= 0.25*(-B[i,j]+sol[iter-1,i+1,j]+sol[iter-1,i-1,j]+sol[iter-1,i,j+1]+sol[iter-1,i,j-1])
    return sol
```

```python
def Gauss_Seidal_2D(sol,B):
    Niter,Nx,Ny=sol.shape
    for iter in range(1,Niter,1):
        for i in range(1,Nx-1,1):
            for j in range(1,Ny-1,1):
                sol[iter,i,j]=0.25*(-B[i,j]+sol[iter-1,i+1,j]+sol[iter,i-1,j]+sol[iter-1,i,j+1]+sol[iter,i,j-1]) #we already have i-1 & j-1
    return sol
```

```python
def SOR_2D(sol,B,omega):
    Niter,Nx,Ny=sol.shape
    for iter in range(1,Niter,1):
        for i in range(1,Nx-1,1):
            for j in range(1,Ny-1,1):
                sol[iter,i,j]=omega*(0.25*(-B[i,j]+sol[iter-1,i+1,j]+sol[iter,i-1,j]+sol[iter-1,i,j+1]+sol[iter,i,j-1]))+(1-omega)*sol[iter-1,i,j]
    return sol
```

```python
def Red_Black_Gauss_Seidal_2D(sol):
    Niter,Nx,Ny=sol.shape
    for iter in range(1,Niter,1):
        for i in range(1,Nx-1,1):
            for j in range(1,Ny-1,1):
                if (i+j)%2==0:
                    sol[iter,i,j]=0.25*(-B[i,j]+sol[iter-1,i+1,j]+sol[iter-1,i-1,j]+sol[iter-1,i,j+1]+sol[iter-1,i,j-1]) 
                
        for i in range(1,Nx-1,1):
            for j in range(1,Ny-1,1):
                if (i+j)%2!=0:
                    sol[iter,i,j]=0.25*(-B[i,j]+sol[iter,i+1,j]+sol[iter,i-1,j]+sol[iter,i,j+1]+sol[iter,i,j-1])
    return sol
```

```python
Nx=100
Ny=100
xmax=10
xmin=0
ymax=10
ymin=0
Niter=400   #if we increase Niter  the peak eventually goes to -1.
omega=0.5
x=np.linspace(0,xmax-xmin,Nx)
y=np.linspace(0,ymax-ymin,Ny)

x_mesh,y_mesh=np.meshgrid(x,y)
sol=np.zeros(((Niter,len(x),len(y))))
sol[0,Nx//2,Ny//2]= 1.0  
  
sol_GS=np.zeros(((Niter,len(x),len(y))))
sol_GS[0,Nx//2,Ny//2]= 1.0

sol_RB_GS=np.zeros(((Niter,len(x),len(y))))
sol_RB_GS[0,Nx//2,Ny//2]= 1.0

sol_SOR=np.zeros(((Niter,len(x),len(y))))
sol_SOR[0,Nx//2,Ny//2]= 1.0
```

```python
xc=x[Nx//2]
yc=y[Ny//2]
dx=x[1]-x[0]
B=np.zeros((len(x),len(y)))
#for i in range(0,Nx,1):
    #for j in range(0,Ny,1):
        #B[i,j]=dx**2*np.exp(-((x[i]-xc)**2+(y[j]-yc)**2))
B=dx**2*np.exp(-((x_mesh-xc)**2+(y_mesh-yc)**2)) # Both of them will work.
```
