RK4

```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
dt=0.25
start_time=0
end_time=50
y_0=5
t_array=np.arange(start_time,end_time,dt)
```

```python
def f(x,y):
    fxy=-y
    return fxy
```

```python
y=np.zeros(np.size(t_array))
k1=np.zeros(np.size(t_array))
k2=np.zeros(np.size(t_array))
k3=np.zeros(np.size(t_array))
k4=np.zeros(np.size(t_array))
y[0]=y_0
```

```python
for t in range(np.size(t_array)-1): #for(t=0;t<40;t++) or for t in range(0,np.size(t_array),1)
    k1[t]=f(t_array[t],y[t])
    k2[t]=f(t_array[t]+(k1[t]*dt/2),y[t])
    k3[t]=f(t_array[t]+(k2[t]*dt/2),y[t])
    k4[t]=f(t_array[t]+(k3[t]*dt),y[t])
    y[t+1]= y[t]+ dt*((k1[t]/6)+((k2[t]+k3[t])/3)+(k4[t]/6))
```

```python
plt.plot(t_array,y,label='RK4')
plt.legend()
plt.xlabel('t')
plt.ylabel('y')
plt.show()
```

```python
y_an=y_0*np.exp(-t_array)
```

```python
plt.plot(t_array,y_an)
```

```python
plt.plot(t_array,y_an,label='analytical')
plt.plot(t_array,y,label='RK4')
plt.legend()
plt.xlabel('t')
plt.ylabel('y')
plt.show()
```
