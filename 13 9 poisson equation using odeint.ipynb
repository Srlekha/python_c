```python
from scipy.integrate import odeint
import numpy as np
```

```python
def abc(rhs,x):
    phi,v=rhs
    drhsdt=[v,np.exp(-x**2)]
    return drhsdt 
```

```python
phi0=[1,0]
x=np.linspace(-10,10,50)
```

```python
sol = odeint(abc, phi0, x)
```

```python
import matplotlib.pyplot as plt
plt.plot(x,sol[:,0],'x',label='phi')
plt.plot(x,sol[:,1],'y',label='v')
plt.legend(loc='best')
plt.xlabel('x')
plt.grid()
plt.show()
```
