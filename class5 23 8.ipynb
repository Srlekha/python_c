```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
dt=0.25
start_time=0
end_time=50
y_0=1
#y_0=0
t_array=np.arange(start_time,end_time,dt)
```

```python
def f(x,y):
    fxy=2*y-3*y**2
    return fxy
```

```python
y=np.zeros(np.size(t_array))
y[0]=y_0
```

```python
for t in range(np.size(t_array)-1):  #for(t=0;t<40;t++) or for t in range(0,np.size(t_array),1)
    y[t+1]= y[t]+ dt*f(t_array[t],y[t])  
```

```python
plt.plot(t_array,y,label='explicit')
plt.legend()
plt.xlabel('t')
plt.ylabel('y')
#plt.ylim([-0.02,2])
plt.show()
```

```python
y_an= 2/((-1*np.exp(-2*t_array))+3)
```

```python
plt.plot(t_array,y_an)
```

```python

plt.plot(t_array,y_an,label='analytical')
plt.plot(t_array,y,label='explicit')
plt.legend()
plt.xlabel('t')
plt.ylabel('y')
plt.show()
```
