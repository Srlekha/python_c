```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
dt=0.001
start_time=-5
end_time=5
y_0=0.00000000002
t_array=np.arange(start_time,end_time,dt)
#print(t_array)
```

```python
def f(x,y):
    fxy=-2*x*y
    return fxy
```

```python
y=np.zeros(np.size(t_array))
y[0]=y_0
```

```python
for t in range(np.size(t_array)-1):  #for(t=0;t<40;t++) or for t in range(0,np.size(t_array),1)
    y[t+1]= y[t]+ dt*f(t_array[t],y[t])
```

```python
plt.plot(t_array,y,label='explicit')
plt.legend()
plt.xlabel('t')
plt.ylabel('y')
plt.show()
```

```python
y_an=1.3*np.exp(-t_array**2)
```

```python
plt.plot(t_array,y_an)
```

```python
plt.plot(t_array,y_an,label='analytical')
plt.plot(t_array,y,label='explicit')
plt.legend()
plt.xlabel('t')
plt.ylabel('y')
plt.show()
```
