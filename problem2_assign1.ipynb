```python
from scipy.integrate import odeint
import numpy as np
```

```python
def assign1_2(rhs,x,a,b):
    y,v=rhs
    drhsdx=[v,-(a/x)*v-(b/x**2)*y]
    return drhsdx 
```

```python
a=1.5
b=-0.5
```

```python
y1=[1,-0.5]
```

```python
x=np.linspace(1,10,101)
```

```python
sol = odeint(assign1_2, y1, x, args=(a,b))
```

```python
import matplotlib.pyplot as plt
plt.plot(x,sol[:,0],'x',label='y(x)')
plt.legend(loc='best')
plt.xlabel('x')
plt.grid()
plt.show()
```

```python
y_an=2/3*(x**(-1))+1/3*(x**(0.5))
```

```python
import matplotlib.pyplot as plt
plt.plot(x,sol[:,0],'x',label='y(x)')
plt.plot(x,y_an,label='analytical')
plt.legend(loc='best')
plt.xlabel('x')
plt.grid()
plt.show()
```
