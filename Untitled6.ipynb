```python
#QUESTION -1
import numpy as np
import matplotlib.pyplot as plt

# Markov transition matrix for given graph
A=np.zeros((10,10))
A[0,2]=1
A[1,0]=0.5
A[1,8]=0.5
A[2,1]=0.5
A[2,6]=0.5
A[4,3]=1
A[5,4]=0.5
A[5,7]=0.5
A[6,5]=0.5
A[7,1]=0.5
A[7,4]=0.5
A[7,9]=1
A[8,0]=0.5
A[8,6]=0.5
A[8,7]=0.5
A[9,5]=0.5
A[9,8]=0.5
```

```python
v=np.random.rand(10,1)
v=v/np.linalg.norm(v,2)
v_data=v
eig_val=0
eig_val_data=[[0]]
k=0
while (np.absolute(eig_val-1))>10**(-5):
  w=np.matmul(A,v)
  v=w/np.linalg.norm(w,2)
  v_data=np.hstack((v_data,v))
  eig_val=np.matmul(np.matmul(np.transpose(v),A),v)
  eig_val_data=np.hstack((eig_val_data,eig_val))

n=np.shape(v_data)[1]
residual=np.zeros((1,n))

for i in range(0,n):
  res=(np.reshape(v,(1,10))-v_data[:,i])
  residual[0,i]=np.linalg.norm(res,2)

x=np.linspace(1,n,n)
plt.plot(x,np.reshape(residual,(n,)))
plt.title('Fig 1: 2 norm residual of eigen vectors')
plt.xlabel('iteration number')
plt.ylabel('two norm residual')
plt.show()
```

```python
succ_res=np.zeros((1,n-1))
for i in range(0,n-1):
  res=v_data[:,i]-v_data[:,i+1]
  succ_res[0,i]=np.linalg.norm(res,2)

x=np.linspace(1,n-1,n-1)
plt.plot(x,np.reshape(succ_res,(n-1,)))
plt.title('Fig 2: 2 norm of successive eigen vectors')
plt.xlabel('iteration number')
plt.ylabel('two norm residual')
plt.show()
```

```python
x=np.linspace(1,n-1,n-1)
plt.plot(x,np.reshape(eig_val_data[:,1:],(n-1,)))
plt.title('Fig 3: Rayleigh quotient')
plt.ylim([0.88,1.1])
plt.show()
```
