```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
dt=0.25
start_time=0
end_time=50
y_0=10
t_array=np.arange(start_time,end_time,dt)
#print(t_array)
```

```python
def f(x,y):
    fxy=-y
    return fxy
```

```python
np.size(t_array)
```

```python
t_array[1]
```

```python
t_array[-2]
```

```python
y=np.zeros(np.size(t_array))
#print(y)
y[0]=y_0
print(y)
```

```python
for t in range(np.size(t_array)-1):  #for(t=0;t<40;t++) or for t in range(0,np.size(t_array),1)
    y[t+1]= y[t]+ dt*f(t_array[t],y[t])
    print(t+1,y[t+1])
```

```python
plt.plot(t_array,y)
```

```python
y_an=y_0*np.exp(-t_array)
```

```python
plt.plot(t_array,y_an)
```

```python
plt.plot(t_array,y,label='explicit')
```

```python
plt.plot(t_array,y_an,label='analytical')
plt.plot(t_array,y,label='explicit')
plt.legend()
plt.xlabel('t')
plt.ylabel('y')
plt.xlim((-5,5))
plt.show()
```

## explicit heun

```python
y_exp_heun=np.zeros(np.size(t_array))
y_exp_heun[0]=10
```

```python
for t in range(np.size(t_array)-1):  
    y_tilde= y_exp_heun[t]+ dt*f(t_array[t],y_exp_heun[t])
    y_exp_heun[t+1]=y_exp_heun[t]+0.5*dt*(f(t_array[t],y_exp_heun[t])+f(t_array,y_tilde))
                                          
```

```python
plt.plot(t_array,y_exp_heun)
```

```python
#%matplotlib notebook
plt.plot(t_array,y_an,label='analytical')
plt.plot(t_array,y,label='explicit')
plt.plot(t_array,y_exp_heun,label='explicit_heun')
plt.legend()
plt.xlabel('t')
plt.ylabel('y')
plt.show()
```

```python
y_imp=np.zeros(np.size(t_array))
y_imp[0]=10
```

```python
for t in range(np.size(t_array)-1):  
    y_imp[t+1]=y_imp[t]/(1+dt)
```

```python
plt.plot(t_array,y_an,label='analytical')
plt.plot(t_array,y,label='explicit')
plt.plot(t_array,y_exp_heun,label='explicit_heun')
plt.plot(t_array,y_imp,label='implicit')

plt.legend()
plt.xlabel('t')
plt.ylabel('y')
plt.show()
```

```python
y_imp_heun=np.zeros(np.size(t_array))
y_imp_heun[0]=10
```

```python
for t in range(np.size(t_array)-1):  
    y_imp_heun[t+1]=y_imp_heun[t]*((2-dt)/(2+dt))
```

```python
%matplotlib notebook
plt.plot(t_array,y_an,label='analytical')
plt.plot(t_array,y,label='explicit')
plt.plot(t_array,y_exp_heun,label='explicit_heun')
plt.plot(t_array,y_imp,label='implicit')
plt.plot(t_array,y_imp_heun,label='implicit_heun')

plt.legend()
plt.xlabel('t')
plt.ylabel('y')
plt.show()
```
