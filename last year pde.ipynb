```python
import numpy as np
import matplotlib.pyplot as plt
from ipynb.fs.full.Thomas_Algo import ThomasAlgorithm, create_tridiagonal
```

```python
def create_matrix_t(y,x,**kwargs):
    args=kwargs["args"]
    delx= args[0]
    dt=args[1]
    Nx= len(x)
    dx=x[1]-x[0]
    a=np.zeros(len(x))
    b=np.zeros(len(x))
    c=np.zeros(len(x))
    d=np.zeros(len(x))
    
    for n in range(1,Nx-1,1):
        D=1+(dx*n)/delx
        alpha=D*dt/(2*(dx**2))
        a[n]=-alpha
        b[n]=1+2*alpha
        c[n]=-alpha
        d[n]=y[n]
    return a,b,c,d
```

```python
Nx=100
xmax=10
xmin=0
delx=xmax-xmin
x=np.linspace(0,xmax-xmin,Nx)
t=np.linspace(0,1,1001)
sol=np.zeros((len(t),len(x)))
sol[0,:]=np.sin((3*np.pi*x)/(xmax-xmin))
sol_CN=np.zeros((len(t),len(x)))
#sol_CN[0,:]=np.sin((3*np.pi*x)/(xmax-xmin))
dx=x[1]-x[0]
dt=t[1]-t[0]
alpha=D*dt/(2*(dx**2))
#sol=ThomasAlgorithm(create_matrix,sol,x,args=(),Left="Dirichlet",Right="Dirichlet")
for time in range(1,len(t),1):
    sol[time]=ThomasAlgorithm(create_matrix_t,sol[time-1,:],x,args=(delx,dt,),Left="Dirichlet",Right="Dirichlet")
    #sol_CN[time]=ThomasAlgorithm(crank_nicholson,sol[time-1,:],x,args=(alpha,),Left="Dirichlet",Right="Dirichlet")
```
