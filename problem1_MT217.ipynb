```python
from scipy.integrate import odeint
import numpy as np
import matplotlib.pyplot as plt
```

```python
y_1=1
x=np.linspace(1,50,10000)
```

```python
def f(x,y):
    #fxy=-(3/4)*((y**4)*(x**2))/((x**3)*(y**3))  # Trying to make it an exact ode
    fxy= -(3/4)*(y/x)
    return fxy
```

```python
sol = odeint(f, y_1, x)
print(sol)
```

```python
plt.plot(x,sol,label='odeint')
```

```python
y_an=x**(-3/4)
```

```python
plt.plot(x,y_an,label='analytical')
plt.plot(x,sol,label='odeint')
plt.legend()
plt.xlabel('x')
plt.ylabel('y')
plt.xlim((1,10))
plt.show()
```
